```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head() # See the first 5 rows
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.pairplot(iris, hue='species')
```

```python
from sklearn.model_selection import train_test_split
```

```python
#train test split
X=iris.drop('species', axis=1)
y=iris['species']
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#training model
from sklearn.tree import DecisionTreeClassifier
classifier= DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

```python
#predicting test data
predTree = classifier.predict(X_test)
predTree
```

```python
#Visualizing Tree
from io import StringIO
import pydotplus
dot_data = StringIO()
from sklearn.tree import export_graphviz
import matplotlib.image as mpimg
filename = "iristree.png"
out=export_graphviz(classifier,feature_names=X.columns, out_file=dot_data, filled=True,  special_characters=True,rotate=False)  
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png(filename)
img = mpimg.imread(filename)
plt.figure(figsize=(100, 200))
plt.imshow(img,interpolation='nearest')
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, predTree))
```
